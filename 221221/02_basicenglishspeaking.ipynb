```python
import requests
from bs4 import BeautifulSoup
```

```python
targetSite = "https://basicenglishspeaking.com/daily-english-conversation-topics/"
```

```python
req = requests.get(targetSite)
print(req)
```

```python
html = req.text
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
subject = [] # 파싱한 대화 주제를 저장할 빈 리스트
divs = soup.findAll('div', {'class':'tcb-flex-col'})
i = 0
for div in divs:
    chapters = div.findAll('a')
    for chapter in chapters:
        i += 1
        print('{:2d}. {}'.format(i, chapter.text))
        subject.append(chapter.text)
```

```python
for i in range(len(subject)):
    print('{:2d}. {}'.format(i + 1 , subject[i]))
```
