워드 클라우드를 그레이톤으로 쵸시하는 함수를 만든다.

```python
# word => 워드클라우드로 표시할 단어
# font_size => 워드클라우드로 표시할 단어의 크기
# position => 워드클라우드로 표시할 단어의 위치
# orientation => 워드클라우드로 표시할 단어의 방향
# random_state => 숫자를 넘겨주면 워드클라우드로 표시되는 단어들이 표시되는 위치가 고정된다.
# hwargs => 워드클라우드로 표시할 단어의 글꼴이 저장된 경로
def gray_color_func(word, font_size, position, orientation, random_state=None, **hwargs):
    # print(word, font_size, position, orientation, random_state, hwargs)
    return 'hsl(0, 0%%, %d%%)' % random.randint(30, 60)
```

```python
plt.figure(figsize=(15, 10))
# 워드클라우드를 그레이스케일로 표시하려면 recolor() 함수의 color_func 속성으로 그레이스케일을 리턴하는 함수를 실행한다.
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=200), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
t = Twitter()
```

```python
# konlpy 패키지는 대한민국 헌법 말뭉치 kolaw와 국화법안 말뭉치 kobill을 제공한다.
# constitution.txt는 파일 데이터 없이 파일 이름만 지정하면 읽어올 수 있다.
ko_con_text = kolaw.open('constitution.txt').read()
ko_con_text
```

```python
# 한글 형태소 분석을 실행한다.
tokens_ko = t.nouns(ko_con_text)
tokens_ko
```

```python
# 불용어를 정의한다.
stop_words = ['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안', '정', '관', '위', '기타', '사항', '경우',
             '이상', '범위', '장', '보장', '거나', '요구']
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
```

```python
# 워드 클라우드로 구현할 단어들의 출현 빈도수를 센다
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(50)
```

```python
# 워드 클라우드로 구현할 단어를 뽑아내서 단어를 key로 하고 출현 빈도수를 value로 하는 딕셔너리를 만든다.
data = dict(ko.vocab().most_common(500))
data
```

```python
# 딕셔너리에 저장된 데이터는
wordcloud = WordCloud(
    font_path='C:\\Windows\\Fonts\\NanumGothicCoding.ttf',
    background_color='white',
    relative_scaling=0.5,
).generate_from_frequencies(data)
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
