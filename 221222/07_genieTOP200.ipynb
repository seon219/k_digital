```python
# 노래 제목 크롤링
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20221222&hh=12&rtm=Y&pg='
header = {'User-agent':'Mozilla/5.0 (Windows NT 6.3; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    req = requests.get(targetSite, headers=header)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    
    titles = soup.findAll('a', {'class':'title ellipsis'})
    
    for i in range(len(titles)):
        print('{:3d}. {}'.format(i + 1 , titles[i].text.strip().split('\n')[0]))
```

```python
# 크롤링한 4페이지 결과를 하나의 리스트에 모은다.
genie = []
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20221222&hh=12&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.3; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    req = requests.get(targetSite, headers=header)
    time.sleep(1)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    
    for j in range(50):
        genie.append(
            (artists[j + 5].text.strip().split('\n')[0], titles[j].text.strip().split('\n')[-1])
        )
print(genie)
```

```python
print(dt.now())
for i in range(200):
    print('{:3d}. {} - {}'.format(i + 1, genie[i][0], genie[i][1]))
```

```python
with open('./output/genieTOP200.txt', 'w', -1, 'utf-8') as file:
    file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
    for i in range(200):
        file.write('{:3d}. {} - {}\n'.format(i + 1, genie[i][0], genie[i][1]))
print('genieTOP200.txt 파일로 쓰기 완료')
```
