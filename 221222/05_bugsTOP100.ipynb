```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
import time
```

```python
print(dt.now())
```

```python
targetSite = "https://music.bugs.co.kr/chart"
req = requests.get(targetSite)
print(req)
html = req.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 노래 제목 크롤링
def getRanking():
    titles = soup.findAll('p', {'class':'title'})
    print(len(titles))
    
    rank = []
    for title in titles:
        names = title.findAll('a')
        for name in names:
            #print(name.text)
            rank.append(name.text)
    return rank
```

```python
ranking = getRanking()
for i in range(len(ranking)):
    print('{:2d}. {}'.format(i + 1 , ranking[i]))
```

```python
# 아티스트 크롤링
artists = soup.findAll('p', {'class':'artist'})
print(len(artists))

for i in range(len(artists)):
    #print('{:3d}. {}'.format(i + 1 , artists[i].text.strip()))
    #print('{:3d}. {}'.format(i + 1 , artists[i].text.split('\n')[1]))
    print('{:3d}. {}'.format(i + 1 , artists[i].text.strip().split('\n')[0]))
```

```python
print(dt.now())
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{:3d}. {} - {}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
file = open('./output/bugTOP100.txt', 'w')

#file.write(str(dt.now()) + '\n')
file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    file.write('{:3d}. {} - {}'.format(i + 1, artist, title) + '\n')
    
file.close()
print('파일에 쓰기 완료')
```

```python
# with 구믄을 사용해서 파일을 open 하면 with 블록의 내용이 모두 실행된 후 자동으로 파일을 닫는다.
with open('./output/bugTOP100_with.txt', 'w') as file:
    file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
    for i in range(100):
        artist = artists[i].text.strip().split('\n')[0]
        title = titles[i].text.strip().split('\n')[0]
        file.write('{:3d}. {} - {}\n'.format(i + 1, artist, title))
# ===== with
print('파일에 쓰기 완료')
```
