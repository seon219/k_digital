```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
import time
```

```python
targetSite = "https://www.melon.com/chart/index.htm"
req = requests.get(targetSite)
print(req)
```

```python
targetSite = "https://www.melon.com/chart/index.htm"
header = {'User-agent':'Mozilla/5.0 (Windows NT 6.3; WOW64; Trident/7.0; rv:11.0) like Gecko'}
req = requests.get(targetSite, headers=header)
print(req)
```

```python
html = req.text
print(html)
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 노래 제목 크롤링
titles = soup.findAll('div', {'class':'rank01'})
print(len(titles))

#for div in divs:
#        titles = div.findAll('a')
#        for title in titles:
#            print(title.text)

for i in range(len(titles)):
    print('{:3d}. {}'.format(i + 1 , titles[i].text.strip().split('\n')[0]))
```

```python
# 아티스트 크롤링
artists = soup.findAll('span', {'class':'checkEllipsis'})
print(len(artists))

for i in range(len(artists)):
    print('{:3d}. {}'.format(i + 1 , artists[i].text.strip().split('\n')[0]))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
file = open('./output/melonTOP100.txt', 'w')

#file.write(str(dt.now()) + '\n')
file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    file.write('{:3d}. {} - {}'.format(i + 1, artist, title) + '\n')
    
file.close()
print('파일에 쓰기 완료')
```

```python
# 텍스트 파일에 저장된 데이터를 읽어서 화면에 출력한다.
try:
    file = open('./output/melonTOP100.txt', 'r')
    lines = file.readlines()
    for line in lines:
        print(line.rstrip())

except FileNotFoundError:
    print('디스크에 파일이 없습니다.')
```
