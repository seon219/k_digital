```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도 위에 사각화 할 수 있다.
# !pip install folium
import folium
```

```python
# folium의 Map() 함수의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 속성으로 지정한 배율을 사용하는 지도를 만든다.
# folium.Map(location=(위도, 경도), zoom_start=배율[, tiles='지도 스타일'])
folium_map = folium.Map(location=(37.5697987, 126.9837167), zoom_start=100)
# folium 지도 객체에서 save() 함수로 작성된 지도 차트를 html 파일로 저장할 수 있다.
folium_map.save('./output/map.html')
folium_map
```

Map() 함수의 tiles 속성으로 지도 스타일을 지정할 수 있다.  
지도 스타일: Open Street Map(기본값), Stamen Terrain, Stamen Toner

```python
folium_map = folium.Map(location=(37.5697987, 126.9837167), zoom_start=100, tiles='Open Street Map')
folium_map
```

```python
folium_map = folium.Map(location=(37.5697987, 126.9837167), zoom_start=100, tiles='Stamen Terrain')
folium_map
```

```python
folium_map = folium.Map(location=(37.5697987, 126.9837167), zoom_start=100, tiles='Stamen Toner')
folium_map
```

```python
folium_map = folium.Map(location=(37.5697987, 126.9837167), zoom_start=15)
# folium의 Marker() 함수의 location 속성으로 지정한 위치에 마커를 만든다. => add_to() 함수로 지도 객체에 추가한다.
# popup 속성으로 Maker를 클릭했을 때 표시할 문자열을 지정할 수 있다.
folium.Marker(location=[37.57777285, 126.9753105], popup='경복궁').add_to(folium_map)
folium.Marker(location=[37.56562954, 126.97535295], popup='덕수궁').add_to(folium_map)
folium.Marker(location=[37.58085492, 126.99216752], popup='창덕궁').add_to(folium_map)
folium.Marker(location=[37.57886153, 126.99305401], popup='창경궁').add_to(folium_map)
folium.Marker(location=[37.57373478,126.99379278], popup='종묘').add_to(folium_map)

# Marker를 클릭했을 떄 팝업되는 말풍선의 크기가 마음에 들지 않으면 Popup() 함수의 max_width 속성으로 변경한다.
popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.Marker(location=[37.56978787,126.98231896], popup=popup).add_to(folium_map)

folium_map.save('./output/map2.html')
folium_map
```

```python
folium_map = folium.Map(location=(37.5697987, 126.9837167), zoom_start=15)

# Marker()라는 함수의 icon 속성에 Icon() 함수로 Marker의 색상과 아이콘 모양을 지정할 수 있다.
# Marker 위에 표시되는 아이콘은 https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph에서 제공하는 아이콘을 표시할 수 있다.

folium.Marker(location=[37.57777285, 126.9753105], popup='경복궁', 
              icon=folium.Icon(color='red', icon='glyphicon-tree-deciduous')).add_to(folium_map)
folium.Marker(location=[37.56562954, 126.97535295], popup='덕수궁', 
              icon=folium.Icon(color='green', icon='ok')).add_to(folium_map)
folium.Marker(location=[37.58085492, 126.99216752], popup='창덕궁', 
              icon=folium.Icon(color='purple', icon='star')).add_to(folium_map)
folium.Marker(location=[37.57886153, 126.99305401], popup='창경궁', 
              icon=folium.Icon(color='orange', icon='heart')).add_to(folium_map)
folium.Marker(location=[37.57373478,126.99379278], popup='종묘',
             icon=folium.Icon(color='pink', icon='tree-conifer')).add_to(folium_map)

popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.Marker(location=[37.56978787,126.98231896], popup=popup,
             icon=folium.Icon(color='beige', icon='glyphicon-flash')).add_to(folium_map)

folium_map.save('./output/map3.html')
folium_map
```

```python
folium_map = folium.Map(location=(37.5697987, 126.9837167), zoom_start=15)

# folium의 CircleMarker() 함수로 원형 마커를 만들 수 있다.
# radius 속성으로 지도에 표시할 원의 반지름을 지정한다.
# color 속성으로 원의 색상, fill_color 속성으로 원의 내부 색상을 지정한다.

folium.CircleMarker(location=[37.57777285, 126.9753105], popup='경복궁', 
              radius=100, color='red', fill_color='blue').add_to(folium_map)

popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.CircleMarker(location=[37.56978787,126.98231896], popup=popup,
             radius=50, color='red', fill_color='blue').add_to(folium_map)

folium_map.save('./output/map4.html')
folium_map
```
