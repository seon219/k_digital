```python
import warnings
warnings.filterwarnings(action='ignore')
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By # selenium 4 버전 'By'를 사용하기 위해 import 한다.
from bs4 import BeautifulSoup
import time
```

from selenium.webdriver.common.by import By를 import 하고 사용한다.  
driver.find_element_by_class_name('') => driver.find_element(By.CLASS_NAME, '')로 변경됨  
driver.find_element_by_id('') => driver.find_element(By.ID, '')로 변경됨  
driver.find_element_by_name('') => driver.find_element(By.NAME, '')로 변경됨  
driver.find_element_by_link_text('') => driver.find_element(By.LINK_TEXT, '')로 변경됨  
driver.find_element_by_partial_link_text('') => driver.find_element(By.PARTIAL_LINK_TEXT, '')로 변경됨  
driver.find_element_by_tag_name('') => driver.find_element(By.TAG_NAME, '')로 변경됨  
driver.find_element_by_css_selector('') => driver.find_element(By.CSS_SELECTOR, '')로 변경됨  
driver.find_element_by_xpath('') => driver.find_element(By.XPATH, '')로 변경됨  

```python
# 셀레니움 옵션
option = Options()
option.add_argument('--window-size=945,1030') # 창 크기(가로, 세로), 픽셀 단위
driver = webdriver.Chrome('./chromedriver.exe', options=option)
driver.implicitly_wait(5)


driver.get('https://comic.naver.com/comment/comment?titleId=796152&no=33')
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
reviews = soup.select('span.u_cbox_contents')
for review in reviews[:1]:
    print(review.text)

    
# 답글 버튼을 클릭한다.
driver.find_element(By.XPATH,
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[1]/div/div[4]/a' # 2번째 댓글의 답글 클릭
).click() 
time.sleep(1)


# selenium 새로고침 방법
# driver.refresh() # 새로고침을 실행
# driver.get(driver.current_url) # get() 함수에 현재 페이지 주소를 넣어주는 방법으로 새로고침을 할 수 있다.
# execute_script() 함수로 자바스크립트 코드를 실행시킬 수 있다.
# driver.execute_script('location.reload()') # 자바 스크립트 location 객체의 reload() 함수를 실행해서 새로고침을 할 수 있다.


# 페이지 내용이 변경되면 다시 읽어와야 한다.
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('배댓 댓글 1페이지=====================================')

driver.find_element(By.XPATH,
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[1]' 
).click() 
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('배댓 댓글 2페이지=====================================')

driver.find_element(By.XPATH,
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[3]' 
).click() 
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('배댓 댓글 3페이지=====================================')

driver.find_element(By.XPATH,
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[4]' 
).click() 
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('배댓 댓글 4페이지=====================================')

driver.find_element(By.XPATH,
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[5]' 
).click() 
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('배댓 댓글 5페이지=====================================')

driver.find_element(By.XPATH,
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[6]' 
).click() 
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('배댓 댓글 6페이지=====================================')

driver.find_element(By.XPATH,
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[7]' 
).click() 
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('배댓 댓글 7페이지=====================================')
```
