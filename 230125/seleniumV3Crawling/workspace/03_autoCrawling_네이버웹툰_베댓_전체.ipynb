```python
import warnings
warnings.filterwarnings(action='ignore')
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
```

```python
option = Options()
option.add_argument('--window-size=945,1030') # Ï∞Ω ÌÅ¨Í∏∞(Í∞ÄÎ°ú, ÏÑ∏Î°ú), ÌîΩÏÖÄ Îã®ÏúÑ
driver = webdriver.Chrome('./chromedriver.exe', options=option)
driver.implicitly_wait(5)


driver.get('https://comic.naver.com/comment/comment?titleId=796152&no=33')
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
reviews = soup.select('span.u_cbox_contents')


# Î≤†ÎåìÏùò Í∞úÏàòÎßåÌÅº Î∞òÎ≥µÌïúÎã§.
for i in range(len(reviews[:])):
    print('{:02d}Î≤àÏß∏ Î≤†Îåì\n{}'.format(i+1, reviews[i].text))
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # Î≤†Îåì ÎãµÍ∏ÄÏùÑ Ïó∞Îã§.
    driver.find_element_by_xpath(
        '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[' + str(i + 1) + ']/div[1]/div/div[4]/a'
    ).click()
    time.sleep(1)

    
    # Î≤†Îåì 1Î≤àÏß∏ ÎãµÍ∏Ä ÌéòÏù¥ÏßÄÎ•º ÏùΩÎäîÎã§.
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
    print('{:d}Î≤àÏß∏ Î∞∞ÎéÉ ÌéòÏù¥ÏßÄ ÎãµÍ∏Ä Í∞úÏàò: {}Í∞ú'.format(i + 1, len(replys)))
    
    
    # Î∞∞ÎéÉ 1Î≤àÏß∏ ÎãµÍ∏Ä ÌéòÏù¥ÏßÄÏóê ÎãµÍ∏ÄÏù¥ ÏûàÏúºÎ©¥ Ï∂úÎ†•ÌïúÎã§.
    if len(replys) == 0:
        print('ÎãµÍ∏Ä ÏóÜÏùå')
    else :
        for reply in replys:
            print(reply.text)
    print()
    
    
    # Î≤†Îåì 1Î≤àÏß∏ ÎãµÍ∏Ä ÌéòÏù¥ÏßÄÎ•º Ï†úÏô∏Ìïú ÎÇòÎ®∏ÏßÄ ÎãµÍ∏Ä ÌéòÏù¥ÏßÄ Í∞úÏàòÎ•º ÏñªÏñ¥Ïò®Îã§.
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    page = soup.select('.u_cbox_reply_area .u_cbox_page')
    print('Î≤†Îåì ÎåìÍ∏Ä ÌéòÏù¥ÏßÄ Í∞úÏàò: {}'.format(len(page)))
    
    #cbox_module_wai_u_cbox_content_wrap_tabpanel > ul > li.u_cbox_comment.cbox_module__comment_447909503._user_id_no_2bzbw8 > div.u_cbox_reply_area > div.u_cbox_paginate > div > a:nth-child(4)
    #cbox_module_wai_u_cbox_content_wrap_tabpanel > ul > li.u_cbox_comment.cbox_module__comment_447909503._user_id_no_2bzbw8 > div.u_cbox_reply_area > div.u_cbox_paginate > div > a:nth-child(5)
    # 1Î≤àÏß∏ Î≤†Îåì ÌéòÏù¥ÏßÄÏùò ÎåìÍ∏ÄÏùÄ ÏùΩÏñ¥ÏôîÏúºÎØÄÎ°ú 2Î≤àÏß∏ Î≤†Îåì ÎåìÍ∏Ä ÌéòÏù¥ÏßÄÎ∂ÄÌÑ∞ ÎßàÏßÄÎßâ Î≤†Îåì ÎåìÍ∏Ä ÌéòÏù¥ÏßÄÍπåÏßÄ Î∞òÎ≥µÌïúÎã§.
    for j in range(1, len(page)):
        # Îã§ÏùåÏóê ÏùΩÏùÑ Î≤†Îåì ÎãµÍ∏ÄÏùÑ Ïó∞Îã§.
        driver.find_element_by_css_selector(
            ' #cbox_module_wai_u_cbox_content_wrap_tabpanel div.u_cbox_reply_area > div.u_cbox_paginate > div > a:nth-child(' + str(j + 3)+ ')'
        ).click()
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
        print(len(replys))
        for reply in replys:
            print(reply.text)
    
    # Î≤†Îåì ÎãµÍ∏ÄÏùÑ Îã´ÎäîÎã§.
    driver.find_element_by_xpath(
        '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[' + str(i + 1) + ']/div[1]/div/div[4]/a'
    ).click()
    time.sleep(1)
```
